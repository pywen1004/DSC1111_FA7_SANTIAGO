```python
import pandas as pd

df = pd.read_csv('senate25-final_updated.csv')

id_vars = df.columns[:12].tolist()
candidate_cols = df.columns[12:].tolist()

df_melted = df.melt(id_vars=id_vars,
                    value_vars=candidate_cols,
                    var_name='Candidate',
                    value_name='Votes')

df_melted['Candidate'] = df_melted['Candidate'].str.replace(r'^\d+\.\s*', '', regex=True)

df_melted.to_csv('senate25_cleaned.csv', index=False)

print("Data transformation complete. Ready for Tableau.")
```
